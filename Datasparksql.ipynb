```python
#create table customer details
create_table_query="""create table customers_details(CustomerKey int,
                                        Gender varchar(200),
                                        Name varchar(200),
                                        City varchar(200),
                                        StateCode varchar(100),
                                        State varchar(200),
                                        ZipCode varchar(100),
                                        Country varchar(200),
                                        Continent varchar(200),
                                        Age int
                                        )"""
cursor.execute(create_table_query)
data="""INSERT INTO customers_details(CustomerKey, Gender, Name, City, StateCode, State, ZipCode, Country, Continent, Age) 
                                        VALUES (%s,%s,%s,%s,%s,%s,%s,%s,%s,%s)"""
for index,row in dfcleancust.iterrows():
    cursor.execute(data,((row['CustomerKey'],
                           row['Gender'],
                           row['Name'],
                           row['City'],
                           row['State Code'],
                           row['State'],
                           row['Zip Code'],
                           row['Country'],
                           row['Continent'],
                           row['Age']))
                    )
connection.commit()
```
