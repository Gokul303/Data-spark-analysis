# Customers

```python
import pandas as pd
df_cust = pd.read_csv(r"C:\Users\ASUS\Desktop\GuviProject2\Customers.csv",encoding='latin1')
df_cust
```

```python
df_cust.info()
```

```python
df_cust.isna().sum()
```

```python
df_cust.fillna(method='ffill',inplace=True)
df_cust.isna().sum()
```

```python
df_cust['Birthday']=pd.to_datetime(df_cust['Birthday'],format="%m/%d/%Y",errors='coerce').dt.year
df_cust.info()
```

```python
df_cust['Age'] = pd.Timestamp.now().year - df_cust['Birthday']
df_cust['Age']
```

```python
df_cust=df_cust.drop(columns=['Birthday'],axis=1)
df_cust
```

```python
# Save the cleaned dataset to a new CSV file
path = r"C:\Users\ASUS\Desktop\GuviProject2\Cleaned_Customers.csv"
df_cust.to_csv(path,index=False)
```

```python
#location
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(14,5))
sns.countplot(x='Country',data=df_cust,palette='viridis',hue='Country')
plt.xlabel("Country")
plt.ylabel("customers")
plt.title("customers as per Country")
```

```python
#Gender
plt.figure(figsize=(14,5))
sns.countplot(x='Gender',data=df_cust,palette='viridis',hue='Gender')
plt.title("Gender")
```

# Products

```python
import pandas as pd
df_Products = pd.read_csv(r"C:\Users\ASUS\Desktop\GuviProject2\Products.csv")
df_Products
```

```python
df_Products.info()
```

```python
df_Products['Unit Cost USD'] = df_Products['Unit Cost USD'].replace({'\$': '', ',': ''}, regex=True)
df_Products['Unit Cost USD'] = pd.to_numeric(df_Products['Unit Cost USD'], errors='coerce')
```

```python
df_Products['Unit Price USD'] = df_Products['Unit Price USD'].replace({'\$': '', ',': ''}, regex=True)
df_Products['Unit Price USD'] = pd.to_numeric(df_Products['Unit Price USD'], errors='coerce')
```

```python
df_Products
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(14,5))
sns.countplot(x='Product Name',data=df_Products,palette='viridis',hue='Product Name')
plt.xlabel("Product Name")
plt.ylabel("Category")
plt.title("Products as per Category")
#hue='Country'
```

```python
path = r"C:\Users\ASUS\Desktop\GuviProject2\Cleaned_Products.csv"
df_Products.to_csv(path,index=False)
```

```python
df_stores['Square Meters'] = df_stores['Square Meters'].round().astype(int)
```
